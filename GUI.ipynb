```python
import tkinter as tk

# Function to handle button click
def button_click():
    print("Button Clicked!")  # This will print in the console

# Create the main window
root = tk.Tk()
root.title("Simple GUI ")

# Create a label
label = tk.Label(root, text="Hello, welcome to GUI programming!")
label.pack()  # Add the label to the window

# Create a button and link it to the function
button = tk.Button(root, text="Click Me", command=button_click)
button.pack()  # Add the button to the window

# Start the GUI event loop
root.mainloop()
```

create database teach;
use teach;
create table student(roll_no int(50),first_name varchar(50),last_name varchar(50),email varchar(100),password varchar(100));
select*from student;

```python
from tkinter import *
import pymysql
import tkinter.messagebox as m

# Create main window
r = Tk()
r.title("Welcome to Yogesh Window")
r.geometry("2000x2000")
r.config(bg="orange")

# Welcome Label
l = Label(r, text="Welcome to Codegyan Services", font="times 30 bold").place(x=300, y=100)

# Function to establish a database connection
def CreateConn():
    return pymysql.connect(host="localhost", database='teach', port=3306, user='root', password='')

# Function to insert data
def InsertData():
    rn = ern.get()
    f = efn.get()
    l = eln.get()
    e = eem.get()
    p = eps.get()
    
    if rn == "" or f == "" or l == "" or e == "" or p == "":
        m.showinfo("Insert status", "All fields are mandatory")
    else:
        try:
            conn = CreateConn()
            cursor = conn.cursor()
            query = "INSERT INTO student (roll_no, first_name, last_name, email, password) VALUES (%s, %s, %s, %s, %s)"
            args = (rn, f, l, e, p)
            cursor.execute(query, args)
            conn.commit()
            m.showinfo("Insert status", "Data inserted successfully")
            conn.close()
        except Exception as ee:
            m.showinfo("Insert status", f"Insert Exception: {ee}")

# Function to search data
def SearchData():
    rn = ern.get()
    
    if rn == "":
        m.showinfo("Search status", "Roll number is mandatory for searching")
    else:
        try:
            conn = CreateConn()
            cursor = conn.cursor()
            query = "SELECT * FROM student WHERE roll_no = %s"
            cursor.execute(query, (rn,))
            row = cursor.fetchone()
            if row:
                efn.delete(0, END)
                eln.delete(0, END)
                eem.delete(0, END)
                eps.delete(0, END)
                
                efn.insert(0, row[1])
                eln.insert(0, row[2])
                eem.insert(0, row[3])
                eps.insert(0, row[4])
                
                m.showinfo("Search status", "Record found")
            else:
                m.showinfo("Search status", "No record found")
            conn.close()
        except Exception as ee:
            m.showinfo("Search status", f"Search Exception: {ee}")

# Function to update data
def UpdateData():
    rn = ern.get()
    f = efn.get()
    l = eln.get()
    e = eem.get()
    p = eps.get()
    
    if rn == "" or f == "" or l == "" or e == "" or p == "":
        m.showinfo("Update status", "All fields are mandatory for updating")
    else:
        try:
            conn = CreateConn()
            cursor = conn.cursor()
            query = "UPDATE student SET first_name=%s, last_name=%s, email=%s, password=%s WHERE roll_no=%s"
            args = (f, l, e, p, rn)
            cursor.execute(query, args)
            conn.commit()
            m.showinfo("Update status", "Data updated successfully")
            conn.close()
        except Exception as ee:
            m.showinfo("Update status", f"Update Exception: {ee}")

# Function to delete data
def DeleteData():
    rn = ern.get()
    
    if rn == "":
        m.showinfo("Delete status", "Roll number is mandatory for deletion")
    else:
        try:
            conn = CreateConn()
            cursor = conn.cursor()
            query = "DELETE FROM student WHERE roll_no = %s"
            cursor.execute(query, (rn,))
            conn.commit()
            efn.delete(0, END)
            eln.delete(0, END)
            eem.delete(0, END)
            eps.delete(0, END)
            ern.delete(0, END)
            m.showinfo("Delete status", "Data deleted successfully")
            conn.close()
        except Exception as ee:
            m.showinfo("Delete status", f"Delete Exception: {ee}")

# Labels
Label(r, text="Roll No", font="times 30 bold").place(x=250, y=200)
Label(r, text="First Name", font="times 30 bold").place(x=250, y=300)
Label(r, text="Last Name", font="times 30 bold").place(x=250, y=400)
Label(r, text="Email", font="times 30 bold").place(x=250, y=500)
Label(r, text="Password", font="times 30 bold").place(x=250, y=600)

# Entry Fields
ern = Entry(r, font="times 30 bold")
ern.place(x=500, y=200)

efn = Entry(r, font="times 30 bold")
efn.place(x=500, y=300)

eln = Entry(r, font="times 30 bold")
eln.place(x=500, y=400)

eem = Entry(r, font="times 30 bold")
eem.place(x=500, y=500)

eps = Entry(r, font="times 30 bold", show="*")
eps.place(x=500, y=600)

# Buttons
Button(r, text="Insert", bg="white", command=InsertData, font="times 30 bold").place(x=200, y=700)
Button(r, text="Search", bg="white", command=SearchData, font="times 30 bold").place(x=400, y=700)
Button(r, text="Update", bg="white", command=UpdateData, font="times 30 bold").place(x=600, y=700)
Button(r, text="Delete", bg="white", command=DeleteData, font="times 30 bold").place(x=800, y=700)

# Main loop to run the GUI
r.mainloop()
```
